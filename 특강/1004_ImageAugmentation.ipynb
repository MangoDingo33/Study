```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image_path = '/Users/mangodingo/Desktop/VS/Study/data/dataset_sample/train/tomato_bacterial_spot/0a1655ed-797c-4d1d-ba35-dc255d68a2ee___GCREC_Bact.Sp 3560.JPG'
image = Image.open(image_path)

plt.imshow(image)
plt.title('Loaded Image')
plt.show()
```

```python
data_transforms = transforms.Compose([transforms.Resize((224, 224)),
                                      transforms.GaussianBlur(kernel_size=(7, 13), sigma=(0.5, 0.5)),
                                      transforms.RandomHorizontalFlip(p=0.5),
                                      transforms.ColorJitter(brightness=(0.7, 1),   #밝기
                                                             contrast = (0.3, 2),   #대비
                                                             saturation = (0.3, 2), #채도
                                                             hue = (-0.01, 0.1)),   #색조
                                      transforms.RandomResizedCrop((224, 224), scale=(0.1, 1), ratio=(0.5, 2)),
                                      transforms.ToTensor(),
                                      transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
```

```python
augmented_image = data_transforms(image)

augmented_image = augmented_image.permute(1, 2, 0)
```

```python
plt.subplot(1,2,1)
plt.imshow(image)
plt.title('Loaded Image')

plt.subplot(1,2,2)
plt.imshow(augmented_image)
plt.title('Augmented Image')

plt.show()
```
